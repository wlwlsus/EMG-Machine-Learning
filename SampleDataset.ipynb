```
import random as r
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive

PI = np.pi
```

```
def MakeExDataSet(count,path,name):
    temp = []
    df_true = []
    df_false = []
    for _ in range(count):
        period = r.uniform(2.9,3.1)
        height = r.uniform(70,80)
        min_y = r.uniform(44,48)
        width_u = r.uniform(2.9,3.1)
        width_d = r.uniform(5,5.3)
        width_range = np.arange(0,2 * PI,0.01)
        Y = abs(height*(1/2)**width_range* np.sin(period*width_range*(width_u/width_d))**2)+ min_y
        Y = np.append(Y, 1)

        temp.append(Y)


    #temp = np.array(temp)
    df_true = pd.DataFrame(temp)

    temp = []

    for j in range(count):
        period = r.uniform(2.9,3.1)
        height = r.uniform(70,80)
        min_y = r.uniform(44,48)
        width_u = r.uniform(2.9,3.1)
        width_d = r.uniform(5,5.3)
        width_range = np.arange(0,2 * PI,0.01)
        
        Y = abs(height*(1/2)**width_range* np.cos(period*width_range*(width_u/width_d))**2)+ min_y
        Y = np.append(Y, 0)

        temp.append(Y)

    #temp = np.array(temp)
    df_false = pd.DataFrame(temp)

    sum_data = pd.concat([df_true,df_false])
    sum_data.to_csv(f'{base_url}/{path}/example_{name}.csv', header=False, index=False)
```

```
MakeExDataSet(15,'data','test')
```
